## Fig.6 (A)

```python
def xgb_fitting(_data):
    
    ntree = 100
    
    cv_xgb = GridSearchCV(
        XGBRegressor(n_estimators=ntree,
                     random_state=929), 
            param_grid={"max_depth": range_depth, 
                    "subsample": range_subsample, 
                    "colsample_bytree": range_colsample,
                    "learning_rate": range_lr})
    
    print('phase-1')
    cv_xgb.fit(_data.iloc[:,:-1], _data.iloc[:,-1])

    cvdepth = cv_xgb.best_params_['max_depth']
    cvsubsample = cv_xgb.best_params_['subsample']
    cvcolsample = cv_xgb.best_params_['colsample_bytree']
    cvlr = cv_xgb.best_params_['learning_rate']

    print('phase-2')
    estimator = XGBRegressor(n_estimators=ntree,\
                       max_depth=cvdepth,\
                       subsample=cvsubsample,\
                       colsample_bytree=cvcolsample,\
                       learning_rate=cvlr,\
                       random_state=929, seed=929, n_jobs=-1)
    best_xgb_ocmdesc = estimator
    
    train_test = shuffle(_data, random_state=10)
    cv = Cv_Pred_Expected(estimator=estimator,\
                      X=train_test.iloc[:,:-1],\
                      y=train_test.iloc[:,-1],\
                      cv=10, redc=True)
    print('phase-3')
    cv.cross_validation()
   
    return cv.importance.mean(axis=1).sort_values(ascending=False)
```

### note on importance types

for recent versions of xgb, you can choose:

```python
itypes = ['weight', 'gain', 'cover', 'total_gain', 'total_cover']
```

here the script uses default xgboost (0.81) importances, but the scores are averaged over cross validation evaluations, which result in more stable results than a one-shot evaluation. However, for careful validation, permutation importances would be more stable (as Breiman mentioned in their CART book). drop-feature importances, randomized importances (oob evaluation) are options.

Either way, be careful about too-strong interpretation. Projecting cross-correlated multivariate phenomena into a single-variable contribution always has information loss after all. This problem is not only for tree ensemble methods, but also for other ML methods. For example, see https://arxiv.org/abs/1710.10547 for neural network cases.

Try multiple methods and importance criteria for more careful assessment.

The model interpretation is one of hot topics in ML. See also papers and textbooks, and the info like
- Beware Default Random Forest Importances https://explained.ai/rf-importance/
- Be careful when interpreting your features importance in XGBoost https://towardsdatascience.com/be-careful-when-interpreting-your-features-importance-in-xgboost-6e16132588e7

Refs:

- The Mythos of Model Interpretability https://arxiv.org/abs/1606.03490
- Interpretable Machine Learning: A Guide for Making Black Box Models Explainable. https://christophm.github.io/interpretable-ml-book/
- 2018 Workshop on Human Interpretability in Machine Learning (WHI) https://sites.google.com/view/whi2018/home

### random experiment

here adding several random features into the data matrix, and see what rank of importances they go. (again, note that the score is averaged over 10-fold cross valiation)

```python
mydata = ocm_desc.copy()
y_keep = mydata['Y(C2),%']
del mydata['Y(C2),%']
```

```python
def run_random_perb(x_data, y_data, rnd_num):
    for i in range(rnd_num):
        x_data['rand-%d' % i] = np.random.random(ocm_desc.shape[0])
    x_data['Y(C2),%'] = y_data
    
    ntree = 100
    
    cv_xgb = GridSearchCV(
        XGBRegressor(n_estimators=ntree,
                     random_state=929), 
            param_grid={"max_depth": range_depth, 
                    "subsample": range_subsample, 
                    "colsample_bytree": range_colsample,
                    "learning_rate": range_lr})
    
    cv_xgb.fit(x_data.iloc[:,:-1],x_data.iloc[:,-1])

    cvdepth = cv_xgb.best_params_['max_depth']
    cvsubsample = cv_xgb.best_params_['subsample']
    cvcolsample = cv_xgb.best_params_['colsample_bytree']
    cvlr = cv_xgb.best_params_['learning_rate']
    
    estimator = XGBRegressor(n_estimators=ntree,\
                       max_depth=cvdepth,\
                       subsample=cvsubsample,\
                       colsample_bytree=cvcolsample,\
                       learning_rate=cvlr,\
                       random_state=929, seed=929, n_jobs=-1)
    best_xgb_ocmdesc = estimator
    
    train_test = shuffle(x_data, random_state=10)
    cv = Cv_Pred_Expected(estimator=estimator,\
                      X=train_test.iloc[:,:-1],\
                      y=train_test.iloc[:,-1],\
                      cv=10, redc=True)
    cv.cross_validation()
    
    rank = cv.importance.mean(axis=1).sort_values(ascending=False)
    r_list = []
    for i in range(rnd_num):
        r = rank.index.get_loc('rand-%d' % i)
        r_list.append(r)
    
    return rank, r_list
```

```python
from collections import defaultdict

num_times = 10
sum_rank = defaultdict(list)
```

For 'weight' (xgboost 0.81 default, and only one choice for python API), adding only 1 random features (uniform distributed) affects a lot. splitting some values at this feature is contributing to decrease the sample errors. (this code base usses xgboost 0.81, and vertion after 0.82 provides selection of many importance types)

But if we add many random features, then selection would have less effects even when we use 'weight'.

```python
x_data = mydata.copy()
rank, _ = run_random_perb(x_data, y_keep, 1)

plt.figure(figsize=(6, 4), dpi=300)
rank.iloc[:20][::-1].plot(kind='barh',color='b')
```

```python
x_data = mydata.copy()
rank, _ = run_random_perb(x_data, y_keep, 100)

plt.figure(figsize=(6, 4), dpi=300)
rank.iloc[:20][::-1].plot(kind='barh',color='b')
```

```python
print(mydata.shape)
num_random_feat = 1

x_data = mydata.copy()
for n in range(num_times):
    print('---%d / %d---' % (n, num_times))
    _, res = run_random_perb(x_data, y_keep, num_random_feat)
    print(np.mean(res), np.min(res), np.max(res))
    sum_rank[num_random_feat] = sum_rank[num_random_feat] + res
```
